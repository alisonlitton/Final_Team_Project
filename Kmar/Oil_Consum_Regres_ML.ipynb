## One-step-ahead Forcast

```python
# Is this running prediction for 2001 - 2010
# Historical prediction 2001 - 2010
# Range will run from 0 to 19
predict0110=[]

for year in range(10):
    start = 1980 + year
    end = 2000 + year

    xy = full_merge.loc[full_merge['Year'].between(start, end), ['World Population', 'World Barrels', 'US Inflation%']]
    
    # Does not need .value.reshape(-1, 1) as there's dimension now with 2 features
    X = xy[["World Population", "US Inflation%"]]
    y = xy["World Barrels"].values.reshape(-1, 1)
#     print (X.iloc[20])
    fitment = model.fit(X, y)
    
    # Do not need to .reshape(-1, 1) in oil predict
    oil_predict = fitment.predict(X.iloc[20].values.reshape(1, -1))
    predict0110.append(oil_predict.flatten()[0])
    
print(predict0110)
    
# predict0110
#     print(start)
#     print(end)
#     print(y)
#     print(f"predict: {oil_predict}")
# 
#     for sublist in oil_predict:
#         for item in sublist:
#             predict0110.append(item)
```

## Multi-step Forecast

```python
multi_list=[]

# example: use 10 years of data (2010 - 2019) for 2020y
# use 2019 to predict 2020y (barrel)
# ex: use 2014 - 2019 avg feature (world pop) for 2020 forecast feature of world pop for x
# use x forecast for rolling window to predict y for 2021
# function in pandas to apply mean to window. append to df (Google how to append single parameter to df)
# in range(5), we can hard code rolling window of year - 5
# If fancy: we can run regression within features to predict features  O__O

for year in range(5):
    # Think about start and end
    start = 2014 + year
    end = 2019 + year

    xy = full_merge.loc[full_merge['Year'].between(start, end), ['World Population', 'World Barrels', 'US Inflation%']]
    
    # Does not need .value.reshape(-1, 1) as there's dimension now with 2 features
    X = xy[["World Population", "US Inflation%"]]
    y = xy["World Barrels"].values.reshape(-1, 1)
#     print (X.iloc[20])
    fitment = model.fit(X, y)
    
    # Do not need to .reshape(-1, 1) in oil predict
    oil_predict = fitment.predict(X.iloc[20].values.reshape(1, -1))
    predict0110.append(oil_predict.flatten()[0])
    
print(predict0110)
```
