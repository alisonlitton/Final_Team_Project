```python
import pandas as pd
import numpy as np
import os
```

```python
inital_df = "../notebook_data/Taxes on goods and services (% of revenue).csv"
```

```python
#Read CVS
df1  = pd.read_csv(inital_df)
df1
```

```python
#Remove characters from dataframe
clean1 = df1.replace('-', np.nan)
clean1
```

```python
#Replace Nan with Zeros from dataframe
clean2 = clean1.replace(np.nan,0)
clean2
```

```python
#Remove Zeros from dataframe
# clean3 = clean2[(clean2 != 0).all(1)]
# clean3
```

```python
#Remove unnecessary columns from dataframe
clean4 = clean2.drop(columns=['Country Code','Indicator Name','Indicator Code'])
clean4
```

```python
# Rename columns
clean5 = clean4.rename(columns = {'Country Name': 'Year'}, inplace = False)
clean5
```

```python
#Transpose data for final file
Final_Clean_Transposed = clean5.set_index("Year").T
Final_Clean_Transposed
```

```python
#Export Final data set
Final_Clean_Transposed.to_csv ('../final_clean_data/Taxes%-YearFixed-Python.csv')
```
